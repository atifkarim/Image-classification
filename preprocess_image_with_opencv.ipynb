```python
import cv2
import glob
import os
import numpy as np
IMG_SIZE=46
```

```python
path = r'/home/atif/training_by_several_learning_process/number_classify/rgb_2_gray/Image-classification/test_image/'
img_path = glob.glob(path+ '/*.ppm')

for image_number,image in enumerate(img_path):
    print("\nName of loaded image: ",image)
    cv_img = []
    n= cv2.imread(image)
    n = cv2.resize(n,(48,48))
    n = cv2.cvtColor(n, cv2.COLOR_BGR2GRAY)
    n = cv2.normalize(n, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
#     n = cv2.normalize(n,  n, 0, 255, cv2.NORM_MINMAX) #1st n source, 2nd n output
    cv_img.append(n)
    cv_img = np.array(cv_img)
    cv_img=cv_img.reshape(48,48)
    cv_img = np.rollaxis(cv_img,-1)
    print(cv_img)
    
#     cv_img = cv_img.reshape(IMG_SIZE,IMG_SIZE)
#     print(cv_img)
#     print("one time finish")
```
